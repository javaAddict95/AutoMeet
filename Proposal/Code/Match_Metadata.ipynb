#Match Metadat

This script is an initial approach to extract match information of different sports, especially football. The API provides a free tier for developers with certainly an up cap limit. The retrieved metadata here provides the following information:

1. Tournament.
2. Stage.
3. Stadium.
4. Status: FT indecats Full Time and HT indecates Half Time.
5. Date.
6. Time.
7. Teams: name, score, red cards, and thumbnail.

Also, other information can be extracted, like the goal time, football possession and other satistics related to the match and the players. This information will be considered at the next stage to be mapped with the football match video.
